```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt
```

Impulso unitario 

```python
def impulso_unitarios(n0,n1,n2):
    n=np.arange(n1,n2+1)    #n=[,]
    x=(n==n0).astype(int)
    return x,n 

x,n=impulso_unitarios(0,-2,2)
print(f'x={x}')
print(f'n={n}')


plt.figure()
plt.stem(n, x, linefmt='b-', markerfmt='bo', basefmt=" ")
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Impulso')
plt.legend()    
plt.grid()
plt.show()
```

Escalón unitario 

```python


def escalon_unitario(n0, n1, n2):
    """
    Genera una secuencia de escalón unitario u[n - n0].
    n0: Punto de inicio del escalón.
    n1, n2: Rango inferior y superior del intervalo.
    """
    # Creamos el vector de tiempo discreto
    n = np.arange(n1, n2 + 1)
    
    # La condición es n >= n0 para que el valor sea 1
    x = (n >= n0).astype(int)
    
    return x, n


# Parámetros: escalón en n0=3, en un rango de -5 a 10
x, n = escalon_unitario(3, -5, 10)

plt.stem(n, x)
plt.title('Función Escalón Unitario $u[n - 3]$')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

Rampa 

```python
def funcion_rampa(n0, n1, n2, m=1):
    """
    Genera una secuencia rampa unitaria o con pendiente m: m * r[n - n0].
    n0: Punto donde inicia la rampa.
    n1, n2: Rango del intervalo.
    m: Pendiente de la rampa (por defecto 1).
    """
    # Creamos el vector de tiempo discreto
    n = np.arange(n1, n2 + 1)
    
    # r[n-n0] = (n - n0) para n >= n0, de lo contrario es 0
    # Usamos np.maximum para asegurar que los valores menores a n0 sean 0
    x = m * (n - n0) * (n >= n0)
    
    return x, n



# Parámetros
m = 2
n0 = 2
x, n = funcion_rampa(n0, -5, 10, m)

plt.stem(n, x)
plt.title(f'Función Rampa con m={m} en $n_0={n0}$')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

Exponencia compleja 


```python
def exponencial_compleja(sigma, omega0, n1, n2):
    """
    Genera una secuencia exponencial compleja x[n] = exp((sigma + j*omega0) * n).
    sigma: Factor de atenuación (negativo) o amplificación (positivo).
    omega0: Frecuencia angular de la oscilación.
    n1, n2: Rango del intervalo.
    """
    # Creamos el vector de tiempo discreto
    n = np.arange(n1, n2 + 1)
    
    # Definimos la señal x[n] = e^((sigma + j*omega0) * n)
    # En Python, la unidad imaginaria 'j' se escribe como 1j
    x = np.exp((sigma + 1j * omega0) * n)
    
    return x, n



# Parámetros: Atenuación leve (-0.1) y una frecuencia de pi/4
sigma = -0.1
omega0 = np.pi / 4
x, n = exponencial_compleja(sigma, omega0, 0, 20)


# Parámetros de ejemplo
sigma = -0.1
omega0 = np.pi / 4
x, n = exponencial_compleja(sigma, omega0, 0, 20)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 7), sharex=True)

# Parte Real - Nota el uso de fr y las dobles llaves {{ }}
ax1.stem(n, x.real)
ax1.set_title(fr'Parte Real $Re\{{x[n]\}}$ (Atenuación $\sigma={sigma}$)')
ax1.grid(True)
ax1.set_ylabel('Amplitud')

# Parte Imaginaria
ax2.stem(n, x.imag, linefmt='r', markerfmt='ro')
ax2.set_title(fr'Parte Imaginaria $Im\{{x[n]\}}$ ($\omega_0={omega0:.2f}$)')
ax2.grid(True)
ax2.set_ylabel('Amplitud')

plt.xlabel('n')
plt.tight_layout()
plt.show()
```

Sinusoidal 

```python
def generar_sinusoide(omega, theta, n1, n2):
    """
    Genera una secuencia sinusoidal: x(n) = sin(omega * n + theta)
    en el intervalo n1 <= n <= n2
    """
    # Generar el vector de tiempo discreto n
    n = np.arange(n1, n2 + 1)
    
    # Calcular la función sinusoidal
    x = np.sin(omega * n + theta)
    
    return n, x

# Ejemplo de uso:
w0 = 0.5   # Frecuencia angular
t0 = 0     # Fase inicial
inicio = 0
fin = 20

n, x = generar_sinusoide(w0, t0, inicio, fin)

# Graficar para visualizar
plt.stem(n, x)
plt.title(f"Sinusoide: x(n) = sin({w0}n + {t0})")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.show()
```

```python

# Definimos el rango total
n = np.arange(0, 18)

# Creamos los tres tramos usando la lógica de r[n-n0] 
# multiplicada por una máscara booleana para limitar el intervalo
x_n = ((n - 0) * ((n >= 0) & (n <= 5)) + 
       (n - 5) * ((n >= 6) & (n <= 11)) + 
       (n - 10) * ((n >= 12) & (n <= 17)))
```

```python
# Calculamos la diferencia finita
# np.diff devuelve un arreglo de longitud N-1, agregamos un 0 para mantener el tamaño
dx_n = np.insert(np.diff(x_n), 0, 0)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Señal Original
ax1.stem(n, x_n)
ax1.set_title('Señal Original $x[n]$ (Rampas por tramos)')
ax1.grid(True)
ax1.set_ylabel('Amplitud')

# Señal Derivada
ax2.stem(n, dx_n, linefmt='r', markerfmt='ro')
ax2.set_title('Derivada Discreta $\Delta x[n]$ (Primera Diferencia)')
ax2.grid(True)
ax2.set_ylabel('Amplitud')

plt.xlabel('n')
plt.tight_layout()
plt.show()
```

Lo que se observa:En las rampas: Mientras la rampa sube con pendiente 1, la derivada es constante e igual a 1.En las transiciones: En los puntos donde la rampa se "reinicia" (por ejemplo, de $n=5$ a $n=6$), se observa un pico negativo. Esto sucede porque el valor cae bruscamente de $5$ a $1$.¿Por qué no es exactamente igual a la derivada continua?En el cálculo continuo, la derivada de una rampa unitaria es un escalón unitario ideal. Sin embargo, en tiempo discreto:Discontinuidades abruptas: En una función continua, la rampa de tu ejercicio tendría saltos verticales. En tiempo discreto, no hay "tiempo" entre $n=5$ y $n=6$. La diferencia finita simplemente resta los dos valores, generando un impulso negativo grande ($1 - 5 = -4$) que no existiría en una derivada continua de funciones suaves.Naturaleza del operador: La derivada continua es un límite diferencial ($\frac{dx}{dt}$), mientras que la discreta es una resta de muestras adyacentes. Esto hace que la derivada discreta sea muy sensible al ruido y a los cambios rápidos de nivel (saltos), convirtiendo cada "caída" del diente de sierra en un pulso negativo.

```python

# Definimos el rango de la señal original basado en el origen en el primer 5
n_orig = np.arange(-5, 18)

# Construcción por partes (Concatenación lógica)
def generar_x_base(n):
    # Rampa ascendente de 0 a 5
    r1 = (n + 5) * ((n >= -5) & (n <= 0))
    # Rampa descendente de 4 a 0
    r2 = (5 - n) * ((n >= 1) & (n <= 5))
    # Rampa ascendente de 1 a 5
    r3 = (n - 5) * ((n >= 6) & (n <= 10))
    # Valores constantes de 5
    s1 = 5 * ((n >= 11) & (n <= 13))
    # Valores constantes de 10
    s2 = 10 * ((n >= 14) & (n <= 17))
    return r1 + r2 + r3 + s1 + s2

x = generar_x_base(n_orig)
```

```python
# Rango extendido para cubrir el desplazamiento
n5 = np.arange(-5, 22)

# Función auxiliar para evaluar x en cualquier n (rellenando con 0 fuera del rango)
def x_eval(n_vec):
    # Aplicamos la lógica de la función base sobre el nuevo vector
    return generar_x_base(n_vec)

x5 = 2 * x_eval(n5 - 4) + x_eval(n5)

plt.figure(figsize=(10, 4))
plt.stem(n5, x5)
plt.title(r'$x_5(n) = 2x(n-4) + x(n)$')
plt.grid(True)
plt.show()
```
