<h3><strong>j)</strong><h3>

```python
#Construcción de la señal y[n] = sen(*pi * 0.12 * n)
n = np.arange(0, 101) #ultimo elemento es 100
y = np.sin(np.pi * 0.12 * n)
```

<h3><strong>k)</strong><h3>

```python
# Construcción de la segunda señal y2[n] = cos(2*pi * 0.03 * n)
y2 = np.cos(2 * np.pi * 0.03 * n)
```

<h3><strong>l)</strong><h3>

```python
# 𝑠[𝑛]=y[n]+𝑦2[𝑛]
s = y + y2
```

```python
#𝑡[𝑛]=y[n].𝑦2[𝑛]
t = y * y2
```

<h3><strong>m)</strong><h3>

```python
plt.figure()
plt.stem(n, y, linefmt='b-', markerfmt='bo', basefmt=" ", label='y[n] = sen(pi * 0.12 * n)')
plt.stem(n, y2, linefmt='r-', markerfmt='ro', basefmt=" ", label='y2[n] = cos(2 * pi * 0.03 * n)')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Señales y[n] y y2[n]')
plt.legend()    
plt.grid()
plt.show()
```

<h3><strong>n)</strong><h3>

```python
plt.figure()
plt.stem(n,s, linefmt='g-', markerfmt='go', basefmt=" ", label='s[n] = y[n] + y2[n]')
plt.stem(n,t, linefmt='m-', markerfmt='mo', basefmt=" ", label='t[n] = y[n] * y2[n]')
plt.xlabel('n')
plt.ylabel('amplitud')
plt.title('Señales s[n] y t[n]')
plt.legend()
plt.grid()
plt.show()
```

```python
dict_ejemplo ={'Maria':2.5, 'Juan':3.0, 'Ana':4.2, 'Luis':1.8, 'Sofia':4.5, 'Carlos':3.7, 'Lucia':4.0, 'Diego':2.9, 'Valentina':4.8, 'Santiago':3.2}

def estadisticas(dict_ejemplo):
    maximo = max(dict_ejemplo.values())
    minimo = min(dict_ejemplo.values())
    desviacion_tipica = np.std(list(dict_ejemplo.values()))
    return maximo, minimo, desviacion_tipica

maximo, minimo, desviacion_tipica = estadisticas(dict_ejemplo)
print(f'Máximo: {maximo}')
print(f'Mínimo: {minimo}')
print(f'Desviación típica: {desviacion_tipica}')

```

```python
def clasificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif imc < 25:
        return "Normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

df["Categoria_IMC"] = df["IMC"].apply(clasificar_imc)
```
